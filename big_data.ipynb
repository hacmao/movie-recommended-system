```python
itemInfo = ml100k.collect()
itemInfo = np.array(itemInfo)
itemFeatures = itemInfo[:, -19:]
print(itemFeatures.shape)
```

```python
# mostSimilarMovie = []
cosineSimilarity = np.zeros([n_items, n_items])
# print(cosineSimilarity)
cosineSimilarity = cosine_similarity(itemFeatures, itemFeatures)
print(cosineSimilarity.shape)

index = np.argsort(cosineSimilarity, axis = 1)
# print(index)
cosineSimilarity = np.take_along_axis(cosineSimilarity, index, axis = 1)
print(cosineSimilarity)

mostSimilarMovie = cosineSimilarity[:, -10:]
# print(mostSimilarMovie.shape)
mostSimilarMovieIndex = index[:, -11:]
print(mostSimilarMovieIndex)
```

```python
def recommendMovie(movieId) : 
    listMovie = mostSimilarMovieIndex[movieId-1]
    # print(listMovie)
    for i in reversed(listMovie):
        if (movieId != i):
            print(itemInfo[i, 1])
```

## Collaborative Filter

### Parameters of ALS Model in PySpark realization are following:

1. NumBlocks is the number of blocks the users and items will be partitioned into in order to parallelize computation.
2. rank is the number of latent factors in the model.
3. maxIter is the maximum number of iterations to run.
4. regParam specifies the regularization parameter in ALS.
5. implicitPrefs specifies whether to use the explicit feedback ALS variant or one adapted for implicit feedback data (defaults to false which means using explicit feedback).
6. alpha is a parameter applicable to the implicit feedback variant of ALS that governs the baseline confidence in preference observations (defaults to 1.0).

```python
%%time
als = ALS(
            rank = 10,
            maxIter = 15, 
            regParam = 0.01, 
            userCol = "userId", 
            itemCol = "movieId", 
            ratingCol = "rating", 
            coldStartStrategy = "drop",
            implicitPrefs = False
        )

model = als.fit(train)
```

```python
%%time
# Evaluate model by computing RMSE on train data 
predictions = model.transform(test)
evaluator = RegressionEvaluator(metricName = 'rmse', labelCol = "rating", predictionCol = "prediction")
RMSE = evaluator.evaluate(predictions)
```

```python
%%time
predictions.show(20)
```

```python
print("Root-mean-square error = " + str(RMSE))
```

```python
%%time
# Generate top 10 movie recommendations for each user
userRecs = model.recommendForAllUsers(10)
userRecs.show()
# Generate top 10 user recommendations for each movie
movieRecs = model.recommendForAllItems(10)
movieRecs.show()
```

## Kafka

Một số hàm cơ bản để xử lý với luồng stream : 
1. `json_load` : phân tích dữ liệu tới dưới dạng json và chuyển về list
2. `getAmount` : trả về số thông tin hiện có trong memory
3. `getNewRecords` : trả về những record mới toiws

```python
from pyspark.sql.functions import col, column
from ast import literal_eval

def json_load(s) :
    try :
        return literal_eval(s.asDict()['value'])
    except Exception as e :
        print(e)
        return None
    
def json_load2(s) :
    try :
        return literal_eval(s)
    except Exception as e :
        print(e)
        return None
    
def getAmount() : 
    return sparkSession.sql("select count(*) from ds")\
            .rdd.map(lambda a : a['count(1)']).collect()[0]

def getNewRecords() : 
    global oldCount
    data = sparkSession.sql("SELECT * from ds")
    res = data.rdd.zipWithIndex().filter(lambda c : c[1] >= oldCount) \
            .map(lambda col: json_load(col[0])) \
            .filter(lambda col : col != None) \
            .collect()
    oldCount = getAmount()
    return res
```

Trong trường hợp này, ta mô phỏng dữ liệu người dùng được truyền từ kafka tới với topic là `messages`. Thực hiện dùng pyspark để lắng nghe luồng stream này. 
Với mỗi 5s, chúng ta sẽ tiến hành ghi ra memory với view là `ds`.

```python
# create new stream subcribe topic "messsages" to receive lastest messages
df = sparkSession.readStream.format("kafka") \
    .option("kafka.bootstrap.servers", "kafka:9092")\
    .option("subscribe", "messages")\
    .option("startingOffsets", """{"messages" : {"0" : -1}}""") \
    .load()

# write to memory each 5s 
ds = df.selectExpr("CAST(value AS STRING)")
ds.writeStream \
    .queryName("ds").format('memory') \
    .trigger(processingTime='5 seconds') \
    .start()
```

Đưa ra gợi ý cho người dùng bằng các dự đoán ngắn hạn, top 10 movie dựa trên movie người dùng vừa thực hiện vote. Với những vote kém, sẽ không đưa ra recommend phim tương tự. 

```python
import time 

oldCount = getAmount()
time.sleep(20)   # wait 20s 
new_record = getNewRecords()
print("[***] Number of new record : %d" % len(new_record))
for record in new_record : 
    print("\n--------------------------------------------------------")
    print("\t[*] Recommend movie for user %d\n" % record['userId'])
    if record['rating'] > 3 : 
        recommendMovie(record['movieId'])
    else : 
        print("\tHappy watching")
```
